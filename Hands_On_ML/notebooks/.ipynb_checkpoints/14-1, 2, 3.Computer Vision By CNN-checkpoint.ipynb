- 시각적 애플리케이션 처리에 중점을 맞춤
    - 다만, 음성인식이나 자연어처리 등의 작업에서도 분명 많이 이용함

# 14.2. 합성곱 층
- CNN에서 가장 중요한 구성유소
- 첫 번째 합성곱 층의 뉴런은 입력 이미지의 모든 픽셀에 연결되는 것이 아니라 합성곱층 뉴런의 수용장 안에 있는 픽셀에만 연결됨
    - 두 번째 합성곱 층에 있는 각 뉴런은 첫 층의 작은 사각영역 안에 위치한 뉴런에 연결됨
    - 이러한 구조는 네트워크가 첫 번째 은닉층에서 작은 저수준 특성에 집중, 그 다음 층에서 더 큰 고수준 특성으로 조합해나갈 수 있도록 조합함
        - 이미지로 표현혀면 아래와 같음
        - 
            - https://sandy-lee.medium.com/how-to-train-neural-networks-for-image-classification-part-2-c9935445c562

- 
- 한 수용장과 다음 수용장 사이 간격을 스트라이드라고 하며, 위 그림은 제로 패딩이 적용된 5 * 7 입력층이 3 * 3수용장과 스트라이드 1을 사용한 것
    - 스트라이드를 조정하여, 모델의 계산 복잡도를 떨어뜨리는 것도 가능하며, 가로, 세로 방향으로 스트라이드가 같아야하는 것은  아님
- 상위 층의 i행, j열에 있는 뉴런이 이전 층에서 i * sh 에서  i * sh + fh-1까지의 행과 j * sw 에서  j * sw + fw - 1까지의 열에 위치한 뉴런과 연결됨
    - sh와 sw는 스트라이드의 수직 값과 수평값을 의미함

### 14.2.1. 필터
- 뉴런의 가중치는 수용장 크기의 작은 이미지로 표현할 수 있음
    - 예를 들어 아래 그림은 **필터**라고 불리는 두 개의 가중치 세트를 보여주고 있음
    - 
        - 첫 번째 것은 가운데 흰 수직선이 있는 검은 사각형 -> 가운데 열은 1로 채워져있고 그 외에는 모두 0인 7 * 7 행렬
            - 즉, 이런 가중치를 사용한 뉴런은 가운데 수직선 부분을 제외하고는 수용장에 있는 모든 것을 무시할 것
        - 두 번째 것은 가운데 흰 수평선이 있는 검은 사각형
            - 여기에서도 이 가중치를 사용한 뉴런은 가운데 수평선 부분을 제외하고는 수용장 내 모든 것을 무시할 것
- 그래서 결국, 하단의 입력 이미지를 네트워크에 주입하고 한 층에 모든 뉴런에 수직선 필터를 적용하면 이 층은 왼쪽 위의 이미지를 출력
    - 비슷하게 오른쪽 위의 이미지는 모든 뉴런에 수평선 필터를 적용한 것
- 이렇게, 층의 뉴런 전체에 적용된 필터 하나는 **특성 맵**하나를 생성함
    - 이 맵은 필터를 가장 극대화시키는 이미지의 영역을 강조함

### 14.2.2. 여러 특성 맵 쌓기
- 실제 합성곱 층은 여러 필터를 가지고 필터마다 하나의 특성 맵을 출력하므로 3D로 표현하는 것이 더 적합함
    - 각 특성 맵의 픽셀은 하나의 뉴런에 해당하고 하나의 특성 맵 안에서는 모든 뉴런이 같은 파라미터(가중치, 편향)을 공유하지만, 다른 특성맵의 뉴런은 다른 파라메터를 사용함
- 그리고 한 뉴런의 수용장은 이전 층에 있는 모든 특성맵에 걸쳐 확장됨
    - 즉, 하나의 합성곱 층이 입력에 여러 필터를 동시에 적용하여 입력에 있는 여러 특성을 연달아 감지할 수 있게 됨

- 입력 이미지는 컬러채널마다 하나씩 여러 서브 층으로 구성되기도 함
    - 컬러 채널은 보통 R, G, B 세 가지 이며, 흑백 이미지는 하나의 채널만 가짐
    - 물론, 매우 많은 채널을 가질 수 있음
        - ex : 위성 이미지는 가시광선 외 다른 빛의 파장도 기록
    - 

- 위 그림을 토대로 구체적으로 보면, 합성곱 층 l에 있는 k특성맵의 (i, j)에 위치한 뉴런은 이전 l-1층에 있는 i * sh에서 i * sh + fh - 1 까지의 행과 i * sw에서  i * sw + fw - 1 까지의 열에 있는 뉴런의 출력에 연결됨
    - 다른 특성 맵이더라도, 같은 i 행과 j 열에 있는 뉴런이라면 정확히 이전 층에 있ㅎ는 동일한 뉴런들의 출력에 연결됨
- 합성곱 층에 있는 뉴런의 출력은 다음과 같이 계산됨
    - z_i,j,k = b_k +SIGMA(u = 0, f_h - 1)SIGMA(v = 0, fw - 1)SIGMA(k'=0, fn' - 1)[x_i',j',k' * W_u,v,k',k]
        - 이 때  i' = i * sh + u, j' = j * sw + v
            - z_i,j,k는 합성곱 층(l층)의 k 특성 맵에서 i, j에 위치한 뉴런의 출력
            - s_h, s_w는 수직, 수평 스트라이드, f_h, f_w는 수용장의 높이와 너비, f_n-1은 이전 층에 있는 특성 맵의 수
            - w_u,v,k',k는 l층의 k특성 맵에 있는 모든 뉴런과 u행, v열, k'특성 맵에 위치한 입력 사이의 연결 가중치

### 14.2.3. 텐서플로 구현
- 각 입력 이미지는 보통 (높, 넢, 채널) 형태의 3D텐서로 표현됨
- 햐나의 미니배치는 (미니배치 크기, 높이, 너비, 채널) 형태의 4D텐서로 표현됨
- 합성곱의 가중는 (fh, fw, fn', fn)형태의 4D텐서로 표현됨
- 편향은 간단하게 (fn)형태의 1D텐서로 나타남

```python
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255
images = np.array([china, flower]) # 이미지 두 개를 이어서 텐서 형태로 만들어줌

batch_size, height, width, channels = images.shape
```

```python
# # 두 개의 필터 만들기
# filters = np.zeros(shape = (7,7, channels, 2), dtype = np.float32)
# filters[:, 3, :, 0] = 1 # 수직선
# filters[3, :, :, 1] = 1 # 수평선

# outputs = tf.nn.conv2d(images, filters, strides = 1, padding = "SAME")

# plt.imshow(outputs[0, :, :, 1], cmap = "gray") # 첫 이미지의 두 번째 특성 맵을 그려봄
# plt.show()
```

- 이유는 모르겠지만 커널이 계속 죽어서 진행이 안되었음
- 코드를 한 줄씩 살펴보자

- 각 컬러 채널의 픽셀 강도는 0 ~ 255사이의 값 -> 따라서 255로 나눠서 스케일링함
- 그 다음 두 개의 7 * 7 필터를 만듦(하나는 가운데 흰 수직선이 있고, 다른 하나는 가운데 흰 수평선이 있음)
- tf.nn.conv2d()함수를 사용하여 위에서 만든 두 필터를 두 이미지에 적용함
- 마지막으로 만들어진 특성 맵 중 하나를 그래프로 그린 것

 - tf.nn.conv2d()
     - images는 입력의 미니배치(4차원) 텐서임
     - filters는 적용될 일련의 필터임
     - strides는 1이나 4개의 원소를 갖는 1D 배열로 지정할 수 있음
         - 1D 배열의 경우 가운데 두 개의 원소는 수직, 수평 스트라이드(sh와 sw)이고, 현재는 첫 번째와 마지막 원소가 1이어야함
         - 배치 스트라이드와 채널 스트레이도 사용될 수 있으나, 일단은 여기서는 수직, 수평 스트라이드만 이용
     - padding은 "VALID"나 "SAME" 중 하나를 지정
         - VALID를 이용할 경우 합성곱 층에 제로패딩을 사용하지 않음
             - 이는 모든 뉴런의 수용장이 입력 안쪽의 영역에만 놓인다는 의미
         - SAME을 이용할 경우, 필요할 때 합성곱 층에 제로 패딩을 사용함
             - 이 경우 출력 크기는 입력 뉴런 수를 스트라이드로 나누어 올림한 것과 같음
                 - ex : 입력값의 크기가 13이고 스트라이드가 5라면 출력의 크기는 3이 됨
             - 필요하면, 가능한한 입력 데이터 주변에 동일하게 0을 추가함

- 보통은 필터를 따로 지정하기 보다는 keras.layers.Conv2D를 이용함

```python
conv = keras.layers.Conv2D(filters = 32, kernel_size = 3, strides = 1,
                          padding = "same", activation = "relu")
```

- 위 코드는 3 * 3 크기의 32개 필터를 이용하며, 스트라이드는 1 적용, 패딩은 SAME적용, 출력을 위해 활성화함수로 ReLU 이용
    - 즉, 합성곱층을 사용하기 위해서는 상당히 다수의 하이퍼파라미터가 필요함
        - 필터의 수, 높이, 너비, 스트라이드, 패딩 종류

### 14.2.4. 메모리 요구사항
- CNN의 가장 큰 문제중 하나가, 합성곱 층이 많은 양의 램을 필요로 한다는 점
    - 특히, 역전파 알고리즘이 역방향 계산할 때 정방향에서 계산했던 모든 중간값을 필요로하기 때문임
- 모델링 이후 추론의 케이스에서는 하나의 층이 점유하고 있는 램은 다음 층의 계산이 완료되자마자 해제될 수 있으므로, 연속된 층에서 필요로 하는만큼의 RAM만 가지고 있으면 되나, 훈련하는 동안에는 모든 값이 보존되어야 함
    - 따라서 적어도 각 층에서 필요한 램 양의 전체 합만큼 필요
- *만약 메모리 부족으로 훈련이 실패할 경우 배치의 크기를 줄여볼 수 있고, 스트라이드를 이용하여 차원을 줄이거나 몇 개 층을 제거할 수도 있음*
